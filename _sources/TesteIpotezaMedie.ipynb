> Problema 1: Evaluarea greutății medii a pungilor de zahăr

Un producător de zahăr dorește să se asigure că greutatea medie a pungilor de zahăr este de 1 kg. Din datele anterioare, se știe că varianța greutății pungilor este de $0,01$ kg². Un eșantion de 40 de pungi este selectat, iar greutatea medie a eșantionului este de 0,98 kg.

Se poate concluziona că greutatea medie a pungilor este diferită de 1 kg? Presupuneți că greutatea pungilor este normal distribuită.



> Problema 2: Verificarea timpului mediu de așteptare într-o clinică

O clinică dorește să mențină timpul mediu de așteptare al pacienților sub 20 de minute. Din studiile anterioare, se știe că varianța timpului de așteptare este de 16 minute². Managerul clinicii măsoară timpul de așteptare pentru un eșantion de 50 de pacienți, obținând o medie de 21,5 minute.

Poate managerul clinicii să concluzioneze că timpul mediu de așteptare depășește 20 de minute? Se consideră că distribuția timpului de așteptare este normală.



> Problema 3: Calitatea producției de suc într-o fabrică

O fabrică de sucuri dorește să se asigure că recipientele sale conțin în medie (exact) 500 ml de suc. Din istoricul producției, se cunoaște că varianța volumului este de $4$ ml². Un eșantion de 30 de recipiente este verificat, iar media eșantionului este de 498 ml.

Este această diferență semnificativă din punct de vedere statistic? Presupuneți că volumul de suc este normal distribuit.



> Problema 4: Evaluarea nivelului mediu de colesterol la o populație

Un medic dorește să determine dacă nivelul mediu de colesterol într-o populație specifică este mai mare de 200 mg/dl. Din datele anterioare, se știe că varianța nivelului de colesterol este de 225 mg²/dl. Medicul testează un eșantion de 100 de persoane și obține un nivel mediu de colesterol de 205 mg/dl.

Poate medicul concluziona că nivelul mediu de colesterol în populație este semnificativ mai mare decât 200 mg/dl? Se presupune că nivelul de colesterol este normal distribuit.



> Problema 5: Timpul mediu de livrare al unui serviciu de curierat

O companie de curierat susține că timpul mediu de livrare a coletelor este de 3 zile. Din datele istorice, compania cunoaște că varianța timpului de livrare este de $0,25$ zile². Un eșantion de 36 de livrări este analizat, iar media eșantionului este de 3,2 zile.

Este această diferență semnificativă? Presupuneți că timpul de livrare este normal distribuit.

> Problema 1: Evaluarea conținutului mediu de cofeină într-o cafea la o cafenea

O cafenea susține că fiecare cafea servită conține în medie 150 mg de cofeină. Un grup de cercetători dorește să verifice această afirmație și colectează un eșantion de 10 cafele. Conținutul mediu de cofeină în eșantion este de 145 mg, iar abaterea medie pătratică  a eșantionului este de 8 mg.

Se poate concluziona că media conținutului de cofeină diferă semnificativ de 150 mg? Se presupune că distribuția conținutului de cofeină este normală.



> Problema 2: Determinarea nivelului mediu de satisfacție a clienților unui hotel

Un hotel de lux susține că nivelul mediu de satisfacție al clienților săi este de 8,5 (pe o scară de la 1 la 10). Pentru a verifica acest lucru, un eșantion de 45 de clienți este intervievat, rezultând o satisfacție medie de 8,2 și o abatere medie pătratică  de 1,1.

Poate fi această diferență considerată semnificativă? Presupuneți că nivelul de satisfacție este normal distribuit.



> Problema 3: Compararea timpului mediu de somn la o populație de studenți

Un studiu anterior a stabilit că studenții unei universități dorm în medie 7 ore pe noapte. Un cercetător dorește să verifice dacă acest timp mediu s-a schimbat și colectează un eșantion de 35 studenți. Media duratei de somn în eșantion este de 6,6 ore, cu o abatere medie pătratică  de 0,9 ore.

Poate cercetătorul concluziona că durata medie de somn este diferită de 7 ore? Se consideră că durata de somn este normal distribuită.



> Problema 4: Calitatea apei potabile într-o zonă rurală

Într-o zonă rurală, concentrația maximă acceptată de nitrați în apa potabilă este de 10 mg/L. Un specialist în sănătate publică suspectează că nivelul mediu de nitrați depășește această valoare și prelevează un eșantion de 12 probe de apă. Media concentrației de nitrați este de 10,5 mg/L, cu o abatere medie pătratică  de 0,6 mg/L.

Există suficiente dovezi pentru a concluziona că nivelul mediu de nitrați este mai mare decât 10 mg/L? Se presupune că nivelul de nitrați este normal distribuit.



> Problema 5: Îmbunătățirea performanței angajaților după un program de instruire

O companie își propune să îmbunătățească performanța angajaților săi, presupunând că aceștia finalizează în mod obișnuit un anumit tip de sarcină în 50 de minute. După implementarea unui nou program de instruire, compania măsoară durata medie necesară pentru a completa sarcina pentru eșantion de 46 angajați, obținând o medie de 47 de minute și o abatere medie pătratică  de 4 minute.

Poate compania concluziona că noul program de instruire a redus semnificativ timpul mediu necesar pentru finalizarea sarcinii? Se presupune că timpii de finalizare a sarcinii sunt normal distribuiți.

> Problema 1: Compararea înălțimii medii între bărbați și femei într-o universitate

Un cercetător dorește să afle dacă există o diferență semnificativă între înălțimea medie a bărbaților și femeilor dintr-o universitate. Eșantioanele sunt selectate astfel:

- **Bărbați**: 40 de studenți, media eșantionului $ \tilde{x}_1 = 175 $ cm, abaterea medie pătratică  $ s_1 = 6 $ cm.
- **Femei**: 38 de studente, media eșantionului $ \tilde{x}_2 = 170 $ cm, abaterea medie pătratică  $ s_2 = 6 $ cm.

Se poate concluziona că există o diferență semnificativă între înălțimea medie a bărbaților și femeilor? Se presupune că înălțimile sunt normal distribuite și au varianțe egale.



> Problema 2: Evaluarea timpului mediu de finalizare a unui test între două grupe de elevi

Un profesor dorește să verifice dacă există o diferență în timpul mediu de finalizare a unui test între două grupe de elevi, după ce aceștia au studiat din surse diferite.

- **Grupa A**: 35 de elevi, timpul mediu de finalizare $ \tilde{x}_1 = 45 $ minute, abaterea medie pătratică  $ s_1 = 5.1 $ minute.
- **Grupa B**: 35 de elevi, timpul mediu de finalizare $ \tilde{x}_2 = 48 $ minute, abaterea medie pătratică  $ s_2 = 4.8 $ minute.

Există o diferență semnificativă între timpii medii de finalizare a testului între cele două grupe? Presupuneți că timpii de finalizare sunt normal distribuiți și au varianțe egale.



> Problema 3: Compararea efectului a două tratamente asupra tensiunii arteriale

Un medic dorește să compare efectul a două tratamente diferite asupra scăderii tensiunii arteriale. Fiecare tratament este administrat unui grup separat de pacienți, iar rezultatele sunt următoarele:

- **Grupul 1**: 15 pacienți, scăderea medie a tensiunii arteriale $ \tilde{x}_1 = 12 $ mmHg, abaterea medie pătratică  $ s_1 = 3 $ mmHg.
- **Grupul 2**: 15 pacienți, scăderea medie a tensiunii arteriale $ \tilde{x}_2 = 10 $ mmHg, abaterea medie pătratică  $ s_2 = 3 $ mmHg.

Poate medicul concluziona că există o diferență semnificativă între efectele celor două tratamente? Se presupune că scăderea tensiunii arteriale este normal distribuită și că ambele grupuri au varianțe egale.



> Problema 4: Salariul mediu lunar al angajaților din două departamente

Poate managerul concluziona că există o diferență semnificativă între salariile medii ale celor două departamente? Se presupune că salariile sunt normal distribuite și au varianțe egale.



> Problema 5: Performanța studenților la un test de matematică în funcție de metoda de predare

Un profesor dorește să compare performanța studenților la un test de matematică între două clase care au fost instruite folosind metode diferite.

- **Metoda tradițională**: 18 studenți, scorul mediu $ \tilde{x}_1 = 75 $, abaterea medie pătratică  $ s_1 = 9 $.
- **Metoda interactivă**: 20 de studenți, scorul mediu $ \tilde{x}_2 = 82 $, abaterea medie pătratică  $ s_2 = 10 $.
